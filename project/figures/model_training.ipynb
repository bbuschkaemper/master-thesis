```python
from csv import DictReader
import matplotlib

matplotlib.use("pgf")
import matplotlib.pyplot as plt


# LaTeX settings
plt.rcParams.update(
    {
        "text.usetex": True,
        "font.family": "EB Garamond",
        "font.size": 11,
        "pgf.texsystem": "lualatex",
    }
)
```

```python
with open("cifar100_overfitting_train.csv", "r") as f:
    reader = DictReader(f)
    steps_train = []
    train = []
    for row in reader:
        steps_train.append(int(row["Step"]))
        train.append(float(row["Value"]))
with open("cifar100_overfitting_val.csv", "r") as f:
    reader = DictReader(f)
    steps_val = []
    val = []
    for row in reader:
        steps_val.append(int(row["Step"]))
        val.append(float(row["Value"]))

# Plotting
plt.figure(figsize=(6, 4))
plt.plot(steps_train, train, label="Train")
plt.plot(steps_val, val, label="Validation")
plt.xlabel("Steps")
plt.ylabel("Accuracy")
plt.title("CIFAR-100 Initial Training Run")
plt.legend()
plt.savefig("../../thesis/figures/cifar100_overfitting.pgf", bbox_inches="tight")
```

```python
import pandas as pd

# Load the eval data
eval_caltech256_2domain_A = pd.read_csv("caltech256_2domain_A_eval.csv")
eval_caltech256_2domain_B = pd.read_csv("caltech256_2domain_B_eval.csv")
eval_caltech256_3domain_A = pd.read_csv("caltech256_3domain_A_eval.csv")
eval_caltech256_3domain_B = pd.read_csv("caltech256_3domain_B_eval.csv")
eval_caltech256_3domain_C = pd.read_csv("caltech256_3domain_C_eval.csv")
eval_cifar100_2domain_A = pd.read_csv("cifar100_2domain_A_eval.csv")
eval_cifar100_2domain_B = pd.read_csv("cifar100_2domain_B_eval.csv")

# Merge the dataframes and add new 'Domain' and 'Dataset Variant' columns
eval_caltech256_2domain_A["Domain"] = "A"
eval_caltech256_2domain_A["Dataset Variant"] = "Caltech-256 2-Domain"
eval_caltech256_2domain_B["Domain"] = "B"
eval_caltech256_2domain_B["Dataset Variant"] = "Caltech-256 2-Domain"
eval_caltech256_3domain_A["Domain"] = "A"
eval_caltech256_3domain_A["Dataset Variant"] = "Caltech-256 3-Domain"
eval_caltech256_3domain_B["Domain"] = "B"
eval_caltech256_3domain_B["Dataset Variant"] = "Caltech-256 3-Domain"
eval_caltech256_3domain_C["Domain"] = "C"
eval_caltech256_3domain_C["Dataset Variant"] = "Caltech-256 3-Domain"
eval_cifar100_2domain_A["Domain"] = "A"
eval_cifar100_2domain_A["Dataset Variant"] = "CIFAR-100 2-Domain"
eval_cifar100_2domain_B["Domain"] = "B"
eval_cifar100_2domain_B["Dataset Variant"] = "CIFAR-100 2-Domain"

# Concatenate all dataframes
eval_data = pd.concat(
    [
        eval_caltech256_2domain_A,
        eval_caltech256_2domain_B,
        eval_caltech256_3domain_A,
        eval_caltech256_3domain_B,
        eval_caltech256_3domain_C,
        eval_cifar100_2domain_A,
        eval_cifar100_2domain_B,
    ],
    ignore_index=True,
)

# Delete 'Wall Time' column
eval_data.drop(columns=["Wall time"], inplace=True)

# Reorder: 'Dataset Variant', 'Domain', 'Step', 'Value'
eval_data = eval_data[["Dataset Variant", "Domain", "Step", "Value"]]

# Make 'Value' a float
eval_data["Value"] = eval_data["Value"].astype(float)

latex_table = eval_data.to_latex(
    index=False,
    caption="Evaluation Results on Test Sets",
    label="tab:evaluation_results",
    column_format="cccc",
    escape=False,
    header=[
        "Dataset Variant",
        "Domain",
        "Steps",
        "Accuracy",
    ],
    bold_rows=True,
    formatters={
        "Value": lambda x: f"{x:.2f}",
    },
)

with open("../../thesis/figures/evaluation_results.tex", "w") as f:
    f.write(latex_table)
```
