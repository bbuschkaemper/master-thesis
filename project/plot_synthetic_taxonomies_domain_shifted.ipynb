```python
import pandas as pd
from csv import DictReader


def calculate_training_duration(file_prefix):
    """Calculate training duration from walltime in training CSV file"""
    try:
        with open(f"training_results/{file_prefix}_train.csv", "r") as f:
            reader = DictReader(f)
            rows = list(reader)

            if not rows:
                return "N/A"

            # Get first and last walltime
            start_time = float(rows[0]["Wall time"])
            end_time = float(rows[-1]["Wall time"])

            # Calculate duration in seconds
            duration_seconds = end_time - start_time

            # Convert to hours and minutes
            hours = int(duration_seconds // 3600)
            minutes = int((duration_seconds % 3600) // 60)

            if hours > 0:
                return f"{hours}h {minutes}m"
            else:
                return f"{minutes}m"
    except FileNotFoundError:
        return "N/A"


# Load the eval data for domain-shifted experiments
eval_caltech256_2domain_domain_shifted_A = pd.read_csv(
    "training_results/caltech256_2domain_domain_shifted_A_eval.csv"
)
eval_caltech256_2domain_domain_shifted_B = pd.read_csv(
    "training_results/caltech256_2domain_domain_shifted_B_eval.csv"
)
eval_caltech256_2domain_variant_domain_shifted_A = pd.read_csv(
    "training_results/caltech256_2domain_variant_domain_shifted_A_eval.csv"
)
eval_caltech256_2domain_variant_domain_shifted_B = pd.read_csv(
    "training_results/caltech256_2domain_variant_domain_shifted_B_eval.csv"
)
eval_caltech256_3domain_domain_shifted_A = pd.read_csv(
    "training_results/caltech256_3domain_domain_shifted_A_eval.csv"
)
eval_caltech256_3domain_domain_shifted_B = pd.read_csv(
    "training_results/caltech256_3domain_domain_shifted_B_eval.csv"
)
eval_caltech256_3domain_domain_shifted_C = pd.read_csv(
    "training_results/caltech256_3domain_domain_shifted_C_eval.csv"
)
eval_cifar100_2domain_domain_shifted_A = pd.read_csv(
    "training_results/cifar100_2domain_domain_shifted_A_eval.csv"
)
eval_cifar100_2domain_domain_shifted_B = pd.read_csv(
    "training_results/cifar100_2domain_domain_shifted_B_eval.csv"
)

# Calculate and add training time to each dataframe
eval_caltech256_2domain_domain_shifted_A["Training Time"] = calculate_training_duration(
    "caltech256_2domain_domain_shifted_A"
)
eval_caltech256_2domain_domain_shifted_B["Training Time"] = calculate_training_duration(
    "caltech256_2domain_domain_shifted_B"
)
eval_caltech256_2domain_variant_domain_shifted_A["Training Time"] = (
    calculate_training_duration("caltech256_2domain_variant_domain_shifted_A")
)
eval_caltech256_2domain_variant_domain_shifted_B["Training Time"] = (
    calculate_training_duration("caltech256_2domain_variant_domain_shifted_B")
)
eval_caltech256_3domain_domain_shifted_A["Training Time"] = calculate_training_duration(
    "caltech256_3domain_domain_shifted_A"
)
eval_caltech256_3domain_domain_shifted_B["Training Time"] = calculate_training_duration(
    "caltech256_3domain_domain_shifted_B"
)
eval_caltech256_3domain_domain_shifted_C["Training Time"] = calculate_training_duration(
    "caltech256_3domain_domain_shifted_C"
)
eval_cifar100_2domain_domain_shifted_A["Training Time"] = calculate_training_duration(
    "cifar100_2domain_domain_shifted_A"
)
eval_cifar100_2domain_domain_shifted_B["Training Time"] = calculate_training_duration(
    "cifar100_2domain_domain_shifted_B"
)

# Merge the dataframes and add new 'Domain' and 'Dataset Variant' columns
eval_caltech256_2domain_domain_shifted_A["Domain"] = "A"
eval_caltech256_2domain_domain_shifted_A["Dataset Variant"] = (
    "Caltech-256 2-Domain Domain Shifted Variant 1"
)
eval_caltech256_2domain_domain_shifted_B["Domain"] = "B"
eval_caltech256_2domain_domain_shifted_B["Dataset Variant"] = (
    "Caltech-256 2-Domain Domain Shifted Variant 1"
)
eval_caltech256_2domain_variant_domain_shifted_A["Domain"] = "A"
eval_caltech256_2domain_variant_domain_shifted_A["Dataset Variant"] = (
    "Caltech-256 2-Domain Domain Shifted Variant 2"
)
eval_caltech256_2domain_variant_domain_shifted_B["Domain"] = "B"
eval_caltech256_2domain_variant_domain_shifted_B["Dataset Variant"] = (
    "Caltech-256 2-Domain Domain Shifted Variant 2"
)
eval_caltech256_3domain_domain_shifted_A["Domain"] = "A"
eval_caltech256_3domain_domain_shifted_A["Dataset Variant"] = (
    "Caltech-256 3-Domain Domain Shifted Variant"
)
eval_caltech256_3domain_domain_shifted_B["Domain"] = "B"
eval_caltech256_3domain_domain_shifted_B["Dataset Variant"] = (
    "Caltech-256 3-Domain Domain Shifted Variant"
)
eval_caltech256_3domain_domain_shifted_C["Domain"] = "C"
eval_caltech256_3domain_domain_shifted_C["Dataset Variant"] = (
    "Caltech-256 3-Domain Domain Shifted Variant"
)
eval_cifar100_2domain_domain_shifted_A["Domain"] = "A"
eval_cifar100_2domain_domain_shifted_A["Dataset Variant"] = (
    "CIFAR-100 2-Domain Domain Shifted Variant"
)
eval_cifar100_2domain_domain_shifted_B["Domain"] = "B"
eval_cifar100_2domain_domain_shifted_B["Dataset Variant"] = (
    "CIFAR-100 2-Domain Domain Shifted Variant"
)

# Concatenate all dataframes
eval_data_domain_shifted = pd.concat(
    [
        eval_caltech256_2domain_domain_shifted_A,
        eval_caltech256_2domain_domain_shifted_B,
        eval_caltech256_2domain_variant_domain_shifted_A,
        eval_caltech256_2domain_variant_domain_shifted_B,
        eval_caltech256_3domain_domain_shifted_A,
        eval_caltech256_3domain_domain_shifted_B,
        eval_caltech256_3domain_domain_shifted_C,
        eval_cifar100_2domain_domain_shifted_A,
        eval_cifar100_2domain_domain_shifted_B,
    ],
    ignore_index=True,
)

# Delete 'Wall Time' and 'Step' columns since we're replacing with Training Time
eval_data_domain_shifted.drop(columns=["Wall time", "Step"], inplace=True)

# Reorder: 'Dataset Variant', 'Domain', 'Training Time', 'Value'
eval_data_domain_shifted = eval_data_domain_shifted[
    ["Dataset Variant", "Domain", "Training Time", "Value"]
]

# Make 'Value' a float
eval_data_domain_shifted["Value"] = eval_data_domain_shifted["Value"].astype(float)

# Rename columns
eval_data_domain_shifted.rename(columns={"Value": "Accuracy"}, inplace=True)

latex_table_domain_shifted = (
    eval_data_domain_shifted.style.hide(axis="index")
    .format(precision=2)
    .to_latex(
        caption="Evaluation results on test sets for domain-shifted experiments. Models were trained with different domain shift transformations and checkpointed after every epoch. The model with the lowest validation loss was selected for evaluation on the test set. Training time indicates the total duration from start to finish of model training.",
        label="tab:evaluation_results_domain_shifted",
        column_format="cccc",
        position="ht",
        position_float="centering",
        hrules=True,
    )
)

with open("../thesis/figures/evaluation_results_domain_shifted.tex", "w") as f:
    f.write(latex_table_domain_shifted)
```

```python
def evaluate_taxonomy(
    constructed_taxonomy, ground_truth_taxonomy, method_name, dataset_name, **params
):
    # Calculate metrics
    edr = constructed_taxonomy.edge_difference_ratio(ground_truth_taxonomy)
    precision, recall, f1 = constructed_taxonomy.precision_recall_f1(
        ground_truth_taxonomy
    )

    # Count relationships for analysis
    num_relationships = len(constructed_taxonomy.graph.edges())
    num_nodes = len(constructed_taxonomy.graph.nodes())

    results = {
        "method": method_name,
        "dataset": dataset_name,
        "edr": edr,
        "precision": precision,
        "recall": recall,
        "f1": f1,
        "num_relationships": num_relationships,
        "num_nodes": num_nodes,
        **params,  # Include method-specific parameters
    }

    return results
```

```python
# Create LaTeX table with best EDR results for domain-shifted experiments
import pandas as pd


# Helper function to find best EDR result for each dataset and method
def find_best_edr_result_domain_shifted(results_dict, dataset):
    if dataset not in results_dict:
        return None

    best_result = None
    best_edr = float("inf")

    for param_value, result in results_dict[dataset].items():
        if result["edr"] < best_edr:
            best_edr = result["edr"]
            best_result = result

    return best_result


# Prepare data for the table with methods as rows
table_data_domain_shifted = []

# Method definitions
methods_domain_shifted = [
    ("MCFP", mcfp_results_domain_shifted, None, "N/A"),
    (
        "Naive Thresholding",
        simple_threshold_results_domain_shifted,
        "threshold",
        "threshold",
    ),
    (
        "Density Thresholding",
        density_threshold_results_domain_shifted,
        "threshold",
        "density_threshold",
    ),
    (
        "Relationship Hypothesis",
        hypothesis_results_domain_shifted,
        "upper_bound",
        "upper_bound",
    ),
]

for dataset, dataset_name in zip(
    datasets_domain_shifted, dataset_titles_domain_shifted
):
    for method_name, results_source, param_key, param_name in methods_domain_shifted:
        if method_name == "MCFP":
            # Handle MCFP results (stored differently)
            mcfp_result = next(
                (r for r in results_source if r["dataset"] == dataset), None
            )
            if mcfp_result:
                table_data_domain_shifted.append(
                    {
                        "Dataset Variant": dataset_name,
                        "Method": method_name,
                        "EDR": f"{mcfp_result['edr']:.3f}",
                        "F1-score": f"{mcfp_result['f1']:.3f}",
                        "Parameter": "N/A",
                    }
                )
        else:
            # Handle parametric methods
            best_result = find_best_edr_result_domain_shifted(results_source, dataset)
            if best_result:
                param_value = best_result[param_key] if param_key else "N/A"
                param_display = (
                    f"{param_value:.2f}"
                    if isinstance(param_value, float)
                    else str(param_value)
                )

                table_data_domain_shifted.append(
                    {
                        "Dataset Variant": dataset_name,
                        "Method": method_name,
                        "EDR": f"{best_result['edr']:.3f}",
                        "F1-score": f"{best_result['f1']:.3f}",
                        "Parameter": param_display,
                    }
                )

# Create DataFrame with simple column structure
df_domain_shifted = pd.DataFrame(table_data_domain_shifted)


# Function to identify best values for each metric within each dataset variant
def get_best_metric_indices_per_dataset_domain_shifted(df):
    """Find the row indices with the best values for each metric within each dataset variant"""
    best_edr_indices = []
    best_f1_indices = []

    # Group by dataset variant and find the best values for each metric
    for dataset_variant in df["Dataset Variant"].unique():
        dataset_rows = df[df["Dataset Variant"] == dataset_variant]

        # Best EDR (lowest value)
        edr_values = dataset_rows["EDR"].astype(float)
        best_edr_idx = edr_values.idxmin()
        best_edr_indices.append(best_edr_idx)

        # Best F1-score (highest value)
        f1_values = dataset_rows["F1-score"].astype(float)
        best_f1_idx = f1_values.idxmax()
        best_f1_indices.append(best_f1_idx)

    return best_edr_indices, best_f1_indices


# Function to identify dataset boundary rows
def get_dataset_boundaries_domain_shifted(df):
    """Identify row indices where dataset variant changes"""
    boundaries = []
    current_dataset = None

    for i, row in df.iterrows():
        dataset_name = row["Dataset Variant"]
        if current_dataset is not None and dataset_name != current_dataset:
            boundaries.append(i)
        current_dataset = dataset_name

    return boundaries


# Get best metric indices and dataset boundaries
best_edr_indices_domain_shifted, best_f1_indices_domain_shifted = (
    get_best_metric_indices_per_dataset_domain_shifted(df_domain_shifted)
)
dataset_boundaries_domain_shifted = get_dataset_boundaries_domain_shifted(
    df_domain_shifted
)


# Function to apply bold formatting to best values independently for each metric
def apply_bold_formatting_domain_shifted(df, best_edr_indices, best_f1_indices):
    """Apply bold formatting to best values for each metric independently"""
    # Create a copy of the dataframe with formatting applied
    formatted_df = df.copy()

    # Bold the best EDR values
    for idx in best_edr_indices:
        formatted_df.loc[idx, "EDR"] = f"\\textbf{{{formatted_df.loc[idx, 'EDR']}}}"

    # Bold the best F1-score values
    for idx in best_f1_indices:
        formatted_df.loc[idx, "F1-score"] = (
            f"\\textbf{{{formatted_df.loc[idx, 'F1-score']}}}"
        )

    return formatted_df


# Apply bold formatting to the DataFrame
df_formatted_domain_shifted = apply_bold_formatting_domain_shifted(
    df_domain_shifted, best_edr_indices_domain_shifted, best_f1_indices_domain_shifted
)

# Create styler with formatted DataFrame
styler_domain_shifted = df_formatted_domain_shifted.style.hide(axis="index")

# Generate LaTeX table without clines first
latex_table_domain_shifted = styler_domain_shifted.to_latex(
    caption="Best EDR results for relationship discovery methods on domain-shifted experiments. For each dataset variant and method, the parameter values that yielded the lowest Edge Difference Ratio (EDR) are shown along with the corresponding F1-score.",
    label="tab:relationship_methods_best_edr_domain_shifted",
    column_format="llccc",  # Left align dataset and method, center align metrics
    position="ht",
    position_float="centering",
    hrules=True,
)


# Post-process LaTeX to add hlines at dataset boundaries
def insert_hlines_at_boundaries_domain_shifted(latex_str, boundaries):
    """Insert \\hline commands at dataset boundaries in LaTeX table"""
    lines = latex_str.split("\n")
    new_lines = []
    data_row_count = 0
    in_table_body = False

    for line in lines:
        # Track when we enter the table body
        if "\\midrule" in line:
            in_table_body = True
            new_lines.append(line)
            continue

        # Check if this is a data row (contains ampersands and ends with \\)
        if in_table_body and " & " in line and line.strip().endswith(" \\\\"):
            # If this row index is a boundary, add hline before it
            if data_row_count in boundaries:
                new_lines.append("\\hline")

            data_row_count += 1

        new_lines.append(line)

    return "\n".join(new_lines)


# Apply the hline insertion
latex_table_domain_shifted = insert_hlines_at_boundaries_domain_shifted(
    latex_table_domain_shifted, dataset_boundaries_domain_shifted
)

# Save to file
with open(
    "../thesis/figures/relationship_methods_results_domain_shifted.tex", "w"
) as f:
    f.write(latex_table_domain_shifted)
```

```python
# Create table with globally optimal parameters and averaged performance metrics for domain-shifted experiments
import pandas as pd


# Function to find globally best parameter for each method across all datasets
def find_globally_best_parameter_domain_shifted(results_dict, param_key, metric="edr"):
    """Find the parameter value that gives the best average performance across all datasets"""
    all_param_values = set()

    # Collect all possible parameter values
    for dataset in results_dict:
        all_param_values.update(results_dict[dataset].keys())

    best_param = None
    best_avg_metric = float("inf") if metric == "edr" else 0

    # For each parameter value, calculate average metric across all datasets
    for param_value in all_param_values:
        metric_values = []

        for dataset in results_dict:
            if param_value in results_dict[dataset]:
                metric_values.append(results_dict[dataset][param_value][metric])

        if metric_values:  # Only consider if we have data for this parameter
            avg_metric = sum(metric_values) / len(metric_values)

            # For EDR, lower is better; for F1, higher is better
            if metric == "edr" and avg_metric < best_avg_metric:
                best_avg_metric = avg_metric
                best_param = param_value
            elif metric == "f1" and avg_metric > best_avg_metric:
                best_avg_metric = avg_metric
                best_param = param_value

    return best_param


# Function to calculate average metrics for a given parameter across all datasets
def calculate_average_metrics_domain_shifted(results_dict, param_value):
    """Calculate average EDR, F1, precision, and recall for a given parameter across all datasets"""
    metrics = {"edr": [], "f1": [], "precision": [], "recall": []}

    for dataset in results_dict:
        if param_value in results_dict[dataset]:
            result = results_dict[dataset][param_value]
            metrics["edr"].append(result["edr"])
            metrics["f1"].append(result["f1"])
            metrics["precision"].append(result["precision"])
            metrics["recall"].append(result["recall"])

    # Calculate averages
    avg_metrics = {}
    for metric_name, values in metrics.items():
        if values:
            avg_metrics[metric_name] = sum(values) / len(values)
        else:
            avg_metrics[metric_name] = None

    return avg_metrics


# Find globally optimal parameters for each method for domain-shifted experiments

# For simple threshold method
best_simple_threshold_domain_shifted = find_globally_best_parameter_domain_shifted(
    simple_threshold_results_domain_shifted, "threshold", "edr"
)
simple_avg_metrics_domain_shifted = calculate_average_metrics_domain_shifted(
    simple_threshold_results_domain_shifted, best_simple_threshold_domain_shifted
)

# For density threshold method
best_density_threshold_domain_shifted = find_globally_best_parameter_domain_shifted(
    density_threshold_results_domain_shifted, "threshold", "edr"
)
density_avg_metrics_domain_shifted = calculate_average_metrics_domain_shifted(
    density_threshold_results_domain_shifted, best_density_threshold_domain_shifted
)

# For hypothesis method
best_upper_bound_domain_shifted = find_globally_best_parameter_domain_shifted(
    hypothesis_results_domain_shifted, "upper_bound", "edr"
)
hypothesis_avg_metrics_domain_shifted = calculate_average_metrics_domain_shifted(
    hypothesis_results_domain_shifted, best_upper_bound_domain_shifted
)

# For MCFP method (no parameters to optimize)
mcfp_metrics_domain_shifted = {"edr": [], "f1": [], "precision": [], "recall": []}
for result in mcfp_results_domain_shifted:
    mcfp_metrics_domain_shifted["edr"].append(result["edr"])
    mcfp_metrics_domain_shifted["f1"].append(result["f1"])
    mcfp_metrics_domain_shifted["precision"].append(result["precision"])
    mcfp_metrics_domain_shifted["recall"].append(result["recall"])

mcfp_avg_metrics_domain_shifted = {}
for metric_name, values in mcfp_metrics_domain_shifted.items():
    mcfp_avg_metrics_domain_shifted[metric_name] = sum(values) / len(values)

# Create table data for globally optimal results
global_table_data_domain_shifted = [
    {
        "Method": "MCFP",
        "Parameter": "N/A",
        "EDR": f"{mcfp_avg_metrics_domain_shifted['edr']:.3f}",
        "Precision": f"{mcfp_avg_metrics_domain_shifted['precision']:.3f}",
        "Recall": f"{mcfp_avg_metrics_domain_shifted['recall']:.3f}",
        "F1-score": f"{mcfp_avg_metrics_domain_shifted['f1']:.3f}",
    },
    {
        "Method": "Naive Thresholding",
        "Parameter": f"{best_simple_threshold_domain_shifted:.2f}",
        "EDR": f"{simple_avg_metrics_domain_shifted['edr']:.3f}",
        "Precision": f"{simple_avg_metrics_domain_shifted['precision']:.3f}",
        "Recall": f"{simple_avg_metrics_domain_shifted['recall']:.3f}",
        "F1-score": f"{simple_avg_metrics_domain_shifted['f1']:.3f}",
    },
    {
        "Method": "Density Thresholding",
        "Parameter": f"{best_density_threshold_domain_shifted:.2f}",
        "EDR": f"{density_avg_metrics_domain_shifted['edr']:.3f}",
        "Precision": f"{density_avg_metrics_domain_shifted['precision']:.3f}",
        "Recall": f"{density_avg_metrics_domain_shifted['recall']:.3f}",
        "F1-score": f"{density_avg_metrics_domain_shifted['f1']:.3f}",
    },
    {
        "Method": "Relationship Hypothesis",
        "Parameter": f"{best_upper_bound_domain_shifted}",
        "EDR": f"{hypothesis_avg_metrics_domain_shifted['edr']:.3f}",
        "Precision": f"{hypothesis_avg_metrics_domain_shifted['precision']:.3f}",
        "Recall": f"{hypothesis_avg_metrics_domain_shifted['recall']:.3f}",
        "F1-score": f"{hypothesis_avg_metrics_domain_shifted['f1']:.3f}",
    },
]

# Create DataFrame
global_df_domain_shifted = pd.DataFrame(global_table_data_domain_shifted)


# Function to find best values for each metric column independently
def apply_bold_to_best_values_domain_shifted(df):
    """Apply bold formatting to the best value in each metric column independently"""
    formatted_df = df.copy()

    # Find best EDR (lowest value)
    edr_values = [float(row["EDR"]) for row in global_table_data_domain_shifted]
    best_edr_value = min(edr_values)

    # Find best Precision (highest value)
    precision_values = [
        float(row["Precision"]) for row in global_table_data_domain_shifted
    ]
    best_precision_value = max(precision_values)

    # Find best Recall (highest value)
    recall_values = [float(row["Recall"]) for row in global_table_data_domain_shifted]
    best_recall_value = max(recall_values)

    # Find best F1-score (highest value)
    f1_values = [float(row["F1-score"]) for row in global_table_data_domain_shifted]
    best_f1_value = max(f1_values)

    # Apply bold formatting to best values
    for idx, row in formatted_df.iterrows():
        if float(row["EDR"]) == best_edr_value:
            formatted_df.loc[idx, "EDR"] = f"\\textbf{{{row['EDR']}}}"

        if float(row["Precision"]) == best_precision_value:
            formatted_df.loc[idx, "Precision"] = f"\\textbf{{{row['Precision']}}}"

        if float(row["Recall"]) == best_recall_value:
            formatted_df.loc[idx, "Recall"] = f"\\textbf{{{row['Recall']}}}"

        if float(row["F1-score"]) == best_f1_value:
            formatted_df.loc[idx, "F1-score"] = f"\\textbf{{{row['F1-score']}}}"

    return formatted_df


# Apply bold formatting to best values in each column
global_df_formatted_domain_shifted = apply_bold_to_best_values_domain_shifted(
    global_df_domain_shifted
)

# Create styler and generate LaTeX table
global_styler_domain_shifted = global_df_formatted_domain_shifted.style.hide(
    axis="index"
)

global_latex_table_domain_shifted = global_styler_domain_shifted.to_latex(
    caption="Average performance metrics for relationship discovery methods with globally optimal parameters on domain-shifted experiments. Each method uses the parameter value that minimizes the average EDR across all domain-shifted dataset variants. Performance metrics are then averaged across all dataset variants using these optimal parameters.",
    label="tab:relationship_methods_global_optimal_domain_shifted",
    column_format="lccccc",  # Left align method, center align others
    position="ht",
    position_float="centering",
    hrules=True,
)

# Save to file
with open(
    "../thesis/figures/relationship_methods_global_optimal_domain_shifted.tex", "w"
) as f:
    f.write(global_latex_table_domain_shifted)
```
